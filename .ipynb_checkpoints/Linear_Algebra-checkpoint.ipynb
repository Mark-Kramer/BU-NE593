# 1. Vectors

```python
import numpy as np

v = np.array([[3], [1], [-7]])
print(v)
```

```python
v.T

print(v.T)
```

```python
vA = np.array([[3, 1, -7]])  # -> 2D row vector (1 row, 3 columns)
print(vA)

vA[0, 1]  #indexing matters! Get (row 0, column 1)
vA[1]
```

```python
vB = np.array([3, 1, -7]) #-> 1D array with 3 elements with no row/col distinction
print(vB)

vB[1]     #indexing matters! Get (second element)
```

# 2. Vector norm

```python
v = np.array([1,7,3,0,1])
```

```python
np.linalg.norm(v)
```

```python
np.sqrt(np.dot(v,v))
```

```python
np.sqrt(v.T @ v)
```

```python
np.sqrt(np.sum(v*v))
```

```python
np.sqrt(np.sum(v**2))
```

# 3. Linear projection

```python
w = np.array([2,2])
v1 = np.array([2,1])
#v1 = np.array([2,0]) 
#v1 = np.array([5,0]) 

# Create the plot
import matplotlib.pyplot as plt
plt.figure()
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(-1, 5)  # Set plot limits a bit beyond the vectors
plt.ylim(-1, 5)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal', adjustable='box')

print(np.dot(w, v1))

print(np.dot(w, v1)/np.sqrt(np.dot(v1,v1)))
```
